```python
import requests
```

```python
res = requests.get("https://docs.python.org/3.5/")
print(res.status_code)
print(res.headers['Content-type'])  # <-- Type of thing that we checked

#print(res.content)  # <-- All content from requests
print(res.text)  # <-- Take all html script
```

```python
res = requests.get("https://docs.python.org/3.5/_static/py.png")
print(res.status_code)
print(res.headers['Content-type'])  # <-- Type of thing that we checked

print(res.content)  # Our picture in code (BINARY REPRESENTATION OF OUR FILE)
# Save that png to file

with open("python.png", "wb") as file:
    file.write(res.content)
```
